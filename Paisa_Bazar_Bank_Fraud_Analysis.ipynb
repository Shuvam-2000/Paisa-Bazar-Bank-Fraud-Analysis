**Checking Outliers**

````python
# plotting a boxplot to check for outliers in Age
plt.figure(figsize=(8, 4))

sns.boxplot(x=data_copy['Age'], color='skyblue', linewidth=1.5)

plt.title("Boxplot of Age", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)

plt.tight_layout()
plt.show()
````

````python
# calculating the IQR and the upper_bound and the lower_bound for age
Q1 = data_copy['Age'].quantile(0.25)
Q3 = data_copy['Age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data_copy[(data_copy['Age'] < lower_bound) | (data_copy['Age'] > upper_bound)]

print("Number of Outliers In Age:", outliers.shape[0])
````

**Observation:- After analyizing with the IQR and boxplot it is observed that the Age variable contains no outliers. Therefore, no outlier treatment was required.**

````python
# calculating the IQR and the upper_bound and the lower_bound for monthly_income
Q1 = data_copy['Monthly_Inhand_Salary'].quantile(0.25)
Q3 = data_copy['Monthly_Inhand_Salary'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data_copy[(data_copy['Monthly_Inhand_Salary'] < lower_bound) | (data_copy['Monthly_Inhand_Salary'] > upper_bound)]

print("Number of Outliers In Monthly_Inhand_Salary:", outliers.shape[0])
````
