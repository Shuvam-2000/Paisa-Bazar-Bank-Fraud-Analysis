## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# making a copy of the dataset to keep the original dataset unchanged
data_copy = data.copy()
````

````python
# checking the copied data info
data_copy.info()
````

````python
# dropping the irrelevant columns not required for analysis
data_copy = data_copy.drop(columns=['ID', 'Name', 'SSN'])
````

````python
data_copy.columns
````

````python
# checking the no of unique values in month column
data_copy['Month'].unique()
````

````python
# fixing incorrect data types
month_map = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug'
}

data_copy['Month'] = data_copy['Month'].map(month_map)
````

````python
data_copy['Month'].unique()
````

````python
# converting the dtype(float to int)
int_columns = [
    'Age',
    'Num_Bank_Accounts',
    'Num_Credit_Card',
    'Interest_Rate',
    'Num_of_Loan',
    'Delay_from_due_date',
    'Num_of_Delayed_Payment',
    'Num_Credit_Inquiries',
    'Credit_History_Age'
]

data_copy[int_columns] = data_copy[int_columns].astype(int)
````
